### -->Discretization

```python
dataset3['dt'] = pd.to_datetime(dataset3['dt'])
dataset3['year'] = dataset3['dt'].map(lambda x: x.year)
dataset3['month'] = dataset3['dt'].map(lambda x: x.month)
```

```python
dataset3['bucket_years'] = pd.cut(x = dataset3['year'], bins = [1740, 1800, 1850, 1900, 1950, 2000, 2013], labels = ['1740', '1800', '1850', '1900', '1950', '2000'])
dataset3['bucket_years']
```

```python
print(dataset3.groupby('bucket_years').size())
```

```python
dataset3['bucket_seasons'] = pd.cut(x = dataset3['month'], bins = [0, 2, 5, 8, 11, 12], labels = ['winter', 'spring', 'summer', 'fall', 'winter'], ordered=False)
dataset3['bucket_seasons']
```

```python
print(dataset3.groupby('bucket_seasons').size())
```

```python
#Drop unnecessary columns
dataset3 = dataset3.drop(['dt'], axis=1)
```

```python
min_year = dataset3['year'].min()
max_year = dataset3['year'].max()
years = range(min_year, max_year+1)
```

```python
#Calculate mean of each season
winter_temp = []
fall_temp = []
summer_temp = []
spring_temp = []

for year in years:
  curr_year = dataset3[dataset3['year'] == year]

  winter_temp.append(curr_year[curr_year['bucket_seasons'] == 'winter']['AverageTemperature'].mean())
  fall_temp.append(curr_year[curr_year['bucket_seasons'] == 'fall']['AverageTemperature'].mean())
  summer_temp.append(curr_year[curr_year['bucket_seasons'] == 'summer']['AverageTemperature'].mean())
  spring_temp.append(curr_year[curr_year['bucket_seasons'] == 'spring']['AverageTemperature'].mean())
```

```python
sns.set(style="whitegrid")
sns.set_color_codes("pastel")
f, ax = plt.subplots(figsize=(10, 6))

plt.plot(years, summer_temp, label='Summers average temperature', color='orange')
plt.plot(years, fall_temp, label='fall average temperature', color='r')
plt.plot(years, spring_temp, label='Springs average temperature', color='g')
plt.plot(years, winter_temp, label='Winters average temperature', color='b')

plt.xlim(min_year, max_year)

ax.set_ylabel('Average temperature')
ax.set_xlabel('Year')
ax.set_title('Average temperature in each season')
legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=True, borderpad=1, borderaxespad=1)
```

## 4.K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

#dataset3['dt'] = pd.to_numeric(pd.to_datetime(dataset3['dt']))
# Create feature and target arrays
X = dataset3.iloc[:, :2]
y = dataset3.bucket_years
```

```python
# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# Predict on dataset which model has not seen before
print(knn.predict(X_test))
print(knn.score(X_test, y_test))
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("\nClassification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
neighbors = np.arange(1,9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

# Loop over K values
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Compute training and test data accuracy
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test)
    
# Generate plot
plt.figure(figsize=(20,10))
plt.plot(neighbors, test_accuracy, label = 'Testing dataset Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training dataset Accuracy')

plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```
